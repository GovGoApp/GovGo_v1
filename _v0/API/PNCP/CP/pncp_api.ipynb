```python
import requests

BASE_URL = "https://pncp.gov.br/api"

def print_relevant(title, response):
    if response.status_code != 200:
        print(f"{title}: Erro {response.status_code}")
        return

    data = response.json()
    print(f"\n--- {title} ---")
    
    # Se a resposta for um dict com chave 'data'
    if isinstance(data, dict) and 'data' in data:
        print("Total de registros:", data.get('totalRegistros'))
        print("Mostrando itens:")
        for item in data['data']:
            # Exibe apenas alguns campos se disponíveis
            num_compra = item.get("numeroCompra", "N/A")
            processo = item.get("processo", "N/A")
            objeto = item.get("objetoCompra", "N/A")
            print(f"Compra: {num_compra} | Processo: {processo} | Objeto: {objeto}")
    # Se a resposta for uma lista
    elif isinstance(data, list):
        for item in data:
            # Se existir o campo 'nome', exibe id e nome
            if "nome" in item:
                ident = item.get("id", "N/A")
                nome = item.get("nome", "N/A")
                print(f"ID: {ident} - Nome: {nome}")
            else:
                print(item)
    else:
        print(data)

# Modalidades de contratação
response = requests.get(f"{BASE_URL}/pncp/v1/modalidades")
print_relevant("Modalidades de contratação", response)

# Instrumentos convocatórios
response = requests.get(f"{BASE_URL}/pncp/v1/tipos-instrumentos-convocatorios")
print_relevant("Instrumentos convocatórios", response)

# Modos de disputa
response = requests.get(f"{BASE_URL}/pncp/v1/modos-disputas")
print_relevant("Modos de disputa", response)

# Categorias Item PCAs
response = requests.get(f"{BASE_URL}/pncp/v1/categoriaItemPcas")
print_relevant("Categorias Item PCAs", response)

# Órgãos (exibe detalhes da unidade e CNPJ do órgão)
response = requests.get(f"{BASE_URL}/pncp/v1/orgaos/27080563000193/unidades")
if response.status_code == 200:
    org_data = response.json()
    print("\n--- Órgãos ---")
    for org in org_data:
        unidade = org.get("nomeUnidade", "N/A")
        cnpj = org.get("orgao", {}).get("cnpj", "N/A")
        print(f"Unidade: {unidade} | CNPJ: {cnpj}")
else:
    print("Órgãos: Erro", response.status_code)

# Catálogos
response = requests.get(f"{BASE_URL}/pncp/v1/catalogos")
if response.status_code == 200:
    cat_data = response.json()
    print("\n--- Catálogos ---")
    for cat in cat_data:
        print(f"ID: {cat.get('id')} | Nome: {cat.get('nome')} | URL: {cat.get('url')}")
else:
    print("Catálogos: Erro", response.status_code)

# Publicações (exibe resumo das contratações)
params = {
    "dataInicial": "20250101",
    "dataFinal": "20250331",
    "codigoModalidadeContratacao": 6,
    "uf": "ES",
    #"palavraChave": "computador",
    "pagina": 2,
    "tamanhoPagina":50
}
response = requests.get(f"{BASE_URL}/consulta/v1/contratacoes/publicacao", params=params)
print_relevant("Publicações", response)
```
